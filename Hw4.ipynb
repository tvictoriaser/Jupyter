## Условие 3: 3 задача

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(12, 10))

sns.set(font_scale=1.4)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix');
```

Наибольшую зависимость цена имеет от кол-ва жилых кв. метров, от качества конструкции здания и дизайна, от кв. метров дома, которые находятся выше земли, от кв. метров жилой площади у 15 соседей, а также от количества ванных комнат.

```python
sns.jointplot(x=df['price'], y=df['floors'], kind='reg');
```

Чем больше в доме этажей, тем выше его стоимость.

```python
sns.boxplot(x=df['price'], y=df['condition'].astype('str'), whis=1.5)

plt.xlabel('price')
plt.ylabel('condition')
plt.title('Distribution of price by condition');
```

Наибольшую стоимость имеют дома с состоянием 5.

```python
plt.figure(figsize=(10, 10))

plt.plot(df['price'], df['sqft_living'])

plt.title('График зависимости цены от жилой площади дома')
plt.xlabel('price')
plt.ylabel('sqft_living');
```

Чем больше жилая площадь дома, тем больше его стоимость.

```python
plt.figure(figsize=(8, 6))

plt.scatter(df['price'], df['grade'])

plt.title('График зависимости цены от качества конструкции здания и дизайна')
plt.xlabel('price')
plt.ylabel('grade');
```

Наибольшую стоимость имеют дома с высокой оценкой качества конструкции здания и дизайна.

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Distribution of price by floors');
```

Наибольшую стоимость имеют дома с количеством этажей 2,5.
